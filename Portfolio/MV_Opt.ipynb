```python
import gurobipy as gp
from gurobipy import GRB, quicksum
import pandas as pd
import copy

import numpy as np
```

```python
df = pd.read_excel('../Naive_S_P500.xlsx', sheet_name='return')
df.head()
```

```python
df.drop(columns=['Unnamed: 0']).head()
```

```python
ret = copy.deepcopy(df.drop(columns=['Unnamed: 0']))
```

```python
ret.head()
```

```python
# iloc[rows, columns]
ret = ret.iloc[:5, 1:397]
n = [i for i in range(396)]
ret
```

```python
cov = ret.cov()
cov.head()
```

```python
var = ret.var()
var.head()
```

```python
var.shape
```

```python
mean = ret.mean()
mean.head()
```

```python
mean[1]
```

```python
mean.iloc[1]
```

```python
m = gp.Model('MV_model')
m.params.NonConvex = 2
```

```python
# decision varibale
w = [m.addVar(lb=0.0, ub=1.0) for i in n]
```

```python
# update
m.update()
```

```python
# objective
eq1 = quicksum(w[i]**2 * var.iloc[i] for i in n) + quicksum([cov.iloc[i, j] * w[i] * w[j] for i in n for j in n if i != j])
m.setObjective(eq1, GRB.MINIMIZE)
```

```python
# set objective
m.setObjective(eq1, GRB.MINIMIZE)
```

```python
# constraints
eq2 = quicksum(mean.iloc[i] * w[i] for i in n)
eq3 = quicksum(w[i] for i in n)
m.addConstr(eq2 >= 0.001)
m.addConstr(eq3 == 1)
```

```python
m.optimize()
for v in m.getVars():
    print('%s %g' % (v.varName, v.x))
print('Obj: %g' % m.objVal)
```
